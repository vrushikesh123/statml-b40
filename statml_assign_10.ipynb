```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/breast_cancer_survival.csv')
data.head()
```

```python
dataf = pd.DataFrame(data)
dataf.head()
```

```python
dataf.dtypes
```

```python
dataf.columns
```

```python
for column in dataf.columns:
    if dataf[column].dtype == 'object':
        unique_values = dataf[column].unique()
        #print(f"Unique values in column '{column}': {unique_values}")
        dataf[column] = dataf[column].map({value: index for index, value in enumerate(unique_values)})
```

```python
dataf.head()
```

```python
x0 = 1
x1 = dataf['Age']
x2 = dataf['Protein1']
x3 = dataf['Protein2']
x4 = dataf['Protein3']
x5 = dataf['Protein4']
x6 = dataf['Tumour_Stage']
x7 = dataf['Histology']
x8 = dataf['ER status']
x9 = dataf['PR status']
x10 = dataf['HER2 status']
x11 = dataf['Surgery_type']
x12 = dataf['Date_of_Surgery']
x13 = dataf['Date_of_Last_Visit']

#For target-Variable.
y = dataf['Patient_Status']
```

```python
x_array = np.array(list(zip(x1, x2, x3, x4, x5, x6, x7, x8, x9, x10, x11, x12, x13)))
y_array = np.array(y)
```

```python
x_array
```

```python
y_array
```

```python
y_array[y_array == '2'] = 0
```

```python
y_array
```

```python
dataf.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

for i in [0.3, 0.4, 0.5]:
    X_train, X_test, y_train, y_test = train_test_split(x_array, y_array, test_size=i, random_state=42)

    # Initialize the KNN model (k = 5)
    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train, y_train)

    # Make prediction
    y_pred = knn.predict(X_test)

    #Now, calculate Accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy for test size {i*100:.0f}%: {accuracy*100:.2f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

test_sizes = [0.2, 0.25, 0.3, 0.35]

for test_size in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(x_array, y_array, test_size=test_size, random_state=42)

    svm = SVC(kernel='linear')
    svm.fit(X_train, y_train)

    #make Predicition.
    y_pred = svm.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy for test size {test_size*100:.0f}%: {accuracy*100:.2f}")
```
